```python
import msprime
import numpy as np
import statistics
import math
import allel
import statsmodels.api as sm
from scipy import stats
import pandas as pd
```

```python

def generate_genotypes(p):
    #where p is the proportion of 'cases'
#     sample_sizes=[2000]*2

#     population_configurations = [
#     msprime.PopulationConfiguration(sample_size=k)
#     for k in sample_sizes]
    
    ncases = int(p*4000)
    ncontrols=int((1-p)*4000)
    population_configurations = [msprime.PopulationConfiguration(sample_size=ncases),
                                msprime.PopulationConfiguration(sample_size=ncontrols)]
    
    

    #specify demographic event - move all lineages to deme d+1 after tmove generations
    demog = [msprime.MassMigration(
            time=20,
            source=0,
            destination=1,
            proportion=1.0)]

    ts=msprime.simulate(Ne=1e4,
                          population_configurations=population_configurations,
                          mutation_rate=1e-08,
                          recombination_rate=1e-08,
                          length=1e6,
                       demographic_events=demog)
    return(ts)
```

```python
#define function to test association between phenotype and a single variant
def assoc(gmat,response):
    #gmat is the genotype matrix
    #response is the response variable
    #variant index is the SNP index
    #predictors=gmat[variant_index,:]
    if(np.mean(gmat)!=0):
        predictors=sm.add_constant(gmat)
        model=sm.OLS(response,predictors).fit()
        beta=model.params[1]
        se=model.bse[1]
        pval=model.pvalues[1]
        #calculate chisq (for calculation of lambda)
        chi=(beta/se)**2
        chi_p=1-stats.chi2.cdf(chi,1)
        results=[beta,se,pval,chi,chi_p]
    else:
        results=[np.nan]*5

    return results
```

```python
def calculate_lambda(tree_sequence,p):
    ht = tree_sequence.genotype_matrix()
    ht = allel.HaplotypeArray(ht)
    daf = np.mean(ht,axis=1)
    cm_ix = np.where(daf>0.05)[0]
    gt = ht[cm_ix,:].to_genotypes(ploidy=2)
    gt2 = gt.to_n_alt()
    phenotype = [[-1]*int(p*2000),[1]*int((1-p)*2000)]
    phenotype2 = [item for sublist in phenotype for item in sublist]
    gwas_results = np.apply_along_axis(assoc,1,gt2,phenotype2)
    gc = np.mean(gwas_results[:,3])
    return(gc)
    
```

```python
ts1 = generate_genotypes(0.1)
ts2 = generate_genotypes(0.2)
ts3 = generate_genotypes(0.3)
ts4 = generate_genotypes(0.4)
ts5 = generate_genotypes(0.5)
```

```python

print(calculate_lambda(ts1,0.1))
print(calculate_lambda(ts2,0.2))
print(calculate_lambda(ts3,0.3))
print(calculate_lambda(ts4,0.4))
print(calculate_lambda(ts5,0.5))
```
